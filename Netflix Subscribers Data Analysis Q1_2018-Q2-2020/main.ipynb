```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
NetflixSubscribersbyCountryfrom2018toQ2_2020_filepath = "NetflixSubscribersbyCountryfrom2018toQ2_2020.csv"

# Read the file into a variable flight_data
Subscribers_data = pd.read_csv(NetflixSubscribersbyCountryfrom2018toQ2_2020_filepath, index_col="Area")
```

```python
# Print the data
Subscribers_data
```

## Heatmap showing Number of subscribers for each Area with each quarter of year

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("Number of subscribers for each Area with time")

# Heatmap Number of subscribers for each Area with each quarter of year
sns.heatmap(data=Subscribers_data, annot=True)

# Add label for horizontal axis
plt.xlabel("Quarter of years from 2018 to Q2-2020")
```

```python
# select numeric columns and calculate the sums
sums = Subscribers_data.select_dtypes(np.number).sum().rename('Total')

# append sums to the data frame
Subscribers_data_Total=Subscribers_data.append(sums)
Subscribers_data_Total
```

```python
# Transposing the dataframe
Subscribers_data_transposed = Subscribers_data_Total.T # Subscribers_data_Total.transpose()
Subscribers_data_transposed
```

## Bar chart showing total Number of subscribers from Q1 2018 to Q2 2020 including all Areas

```python
# Set the width and height of the figure
sns.set_style("whitegrid")
plt.figure(figsize=(14,6))

# Add title
plt.title("Total Number of subscribers from Q1 2018 to Q2 2020 including all Areas", fontname='Comic Sans MS', fontsize=18)

# Bar chart showing total Number of subscribers from Q1 2018 to Q2 2020 including all Areas
sns.barplot(x=Subscribers_data_transposed.index, y=Subscribers_data_transposed['Total'])

# Add label for y axis
plt.ylabel("Number of subscribers", fontname='Arial', fontsize=15)

# Add label for x axis
plt.xlabel("Years", fontname='Arial', fontsize=15)
```

## Now we will show the distribution of subscribers across Different Areas with the help of Pie Charts

```python
Subscribers_data #original dataframe for reference purpose
```

```python
# Extract Column Names as List in Pandas Dataframe
Subscribers_data_columns =list( Subscribers_data.index) 
my_labels = Subscribers_data_columns
my_labels
```

```python
fig = plt.figure(figsize=(18,10))
fig.suptitle("PIE CHARTS SHOWING DISTRIBUTION OF SUBSCRIBERS IN 2018", fontname='DejaVu Sans' ,fontsize=22)


# Plotting the Q1-2018 Pie Chart
ax1 = plt.subplot2grid((2,2),(0,0))
plt.pie(Subscribers_data['Q1 - 2018'] ,labels=my_labels,autopct='%1.1f%%', textprops={'fontsize': 13.5})
plt.title('Distribution of subscribers as of Q1-2018', fontname='Comic Sans MS', fontsize=18)
plt.axis('equal')



# Plotting the  Q2-2018 Pie Chart
ax1 = plt.subplot2grid((2,2),(0,1))
plt.pie(Subscribers_data['Q2 - 2018'] ,labels=my_labels,autopct='%1.1f%%', textprops={'fontsize': 13.5})
plt.title('Distribution of subscribers as of Q2-2018', fontname='Comic Sans MS', fontsize=18)
plt.axis('equal')



# Plotting the  Q3-2018 Pie Chart
ax1 = plt.subplot2grid((2,2),(1,0))
plt.pie(Subscribers_data['Q3 - 2018'] ,labels=my_labels,autopct='%1.1f%%', textprops={'fontsize': 13.5})
plt.title('Distribution of subscribers as of Q3-2018', fontname='Comic Sans MS', fontsize=18)
plt.axis('equal')



# Plotting the  Q4-2018 Pie Chart
ax1 = plt.subplot2grid((2,2),(1,1))
plt.pie(Subscribers_data['Q4 - 2018'] ,labels=my_labels,autopct='%1.1f%%', textprops={'fontsize': 13.5})
plt.title('Distribution of subscribers as of Q4-2018', fontname='Comic Sans MS', fontsize=18)
plt.axis('equal')
```
